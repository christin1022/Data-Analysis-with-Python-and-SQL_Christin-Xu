This is a very popular introductory coding exercise that sometimes shows up in coding interviews for entry-level positions. It's called FizzBuzz. 

Your task: 

1. Write code that takes as input an integer and, for every number from 1 to that integer (inclusive), prints out:
    * If the number is a multiple of both 3 and 5, print "fizzbuzz"
    * If the number is a multiple of 3, print "fizz"
    * If the number is a multiple of 5, print "buzz"
    * Otherwise, print the number
<br></br>
 
2. Identify at least 4 different components of the code; give their name and their definition
    * E.g. Variables, which are names that hold data in a computer program 

```python
def fizzbuzz(limit: int) -> None:
    for num in range(1, limit + 1):
        if num % 3 == 0 and num % 5 == 0:
            print("fizzbuzz")
        elif num % 3 == 0:
            print("fizz")
        elif num % 5 == 0:
            print("buzz")
        else:
            print(num)
```

```python
n = int(input("Enter a positive integer: "))
if n <= 0:
    print("Please enter an integer greater than 0.")
else:
    fizzbuzz(n)
```
